def sentiment_analysis( año : int ): Según el año de lanzamiento, se devuelve una lista con la cantidad de registros de reseñas de usuarios que se encuentren categorizados con un análisis de sentimiento.
Ejemplo de retorno: {Negative = 182, Neutral = 120, Positive = 278}


```python
# Importamos libreria
import pandas as pd
```

```python
# Trabajamos con el df de games_final y reviews_final
df_games = pd.read_csv('./Archivos_csv/games_final.csv')
df_sentiment = pd.read_csv('./Archivos_csv/reviews_final.csv')
```

```python
# Vemos si trae bien cada df
df_games.head()
```

```python
df_sentiment.head()
```

```python
# Hacemos un merge de los data frame por medio de id y de item_id
data_sentiment = df_games.merge(df_sentiment, left_on="id", right_on="item_id", how="inner")
```

```python
# Vemos como queda el DF
data_sentiment.head()
```

```python
data_sentiment.info()
```

```python
# Eliminamos las columnas que no vamos a utilizar, nos quedamos con la columna de year_x que es la de año de relanzamiento

data_sentiment = data_sentiment.drop(columns=['genres', 'app_name', 'early_access', 'id', 'item_id', 'review', 'year_y'])
```

```python
data_sentiment.head()
```

```python
# Elegimos nuevamente los anios con los que  vamos a trabajar, decimios elegir los mismos anios que en la funcion de playTimeGenre y userForGenre
anios = ['2015', '2014', '2013', '2012', '2016', '2011', '2017', '2010', '2009', '2006', '2008', '2007', '2004', '2003', '2005']
```

```python
# Nos quedamos con los años seleccionados
data_sentiment = data_sentiment[data_sentiment['year_x'].isin(anios)]
```

```python
# Vemos como queda la info del DF
data_sentiment.info()
```

```python
data_sentiment.head()
```

```python
# Pasamos el tipo de dato year a int para que nos sirva en la funcion
data_sentiment['year_x'] = data_sentiment['year_x'].astype(int)
```

```python
data_sentiment.info()
```

```python
# Crear variables dummies para "sentiment_analysis"
sentiment_dummies = pd.get_dummies(data_sentiment['sentiment_analysis'], prefix='sentiment', dtype=int)

# Concatenar reviews_df con sentiment_dummies
data_sentiment = pd.concat([data_sentiment, sentiment_dummies], axis=1)
```

```python
data_sentiment.head()
```

```python
# Resetear el indice para que "item_id" pase a columna nuevamente.
data_sentiment.reset_index(inplace=True)
```

```python
data_sentiment.head()
```

```python
data_sentiment = data_sentiment.drop(columns=['title', 'user_id', 'recommend', 'sentiment_analysis'])
```

```python
data_sentiment.info()
```

```python
# Agrupar por 'release_year' y sumar las columnas de análisis de sentimientos
data_sentiment_grouped = data_sentiment.groupby('year_x')[['sentiment_0', 'sentiment_1', 'sentiment_2']].sum().reset_index()

```

```python
# Renombrar las columnas para mayor claridad
data_sentiment_grouped = data_sentiment_grouped.rename(columns={
    'sentiment_0': 'total_negativos',
    'sentiment_1': 'total_neutrales',
    'sentiment_2': 'total_positivos'
})
```

```python
data_sentiment_grouped.head()
```

```python
def sentiment_analysis(year):
    filtro_data = data_sentiment_grouped[data_sentiment_grouped['year_x'] == year]
    
    if filtro_data.empty:
        return {"Negative": 0, "Neutral": 0, "Positive": 0}
    else:
        # Encontrar los valores de sentimiento para el año pasado como parámetro.
        negativo = data_sentiment_grouped.loc[data_sentiment_grouped.year_x == year, "total_negativos"].item()
        neutral = data_sentiment_grouped.loc[data_sentiment_grouped.year_x == year, "total_neutrales"].item()
        positivo = data_sentiment_grouped.loc[data_sentiment_grouped.year_x == year, "total_positivos"].item()
  
    return {"Negative" : negativo, "Neutral" : neutral, "Positive" : positivo}
```

```python
sentiment_analysis(2007)
```

```python
# Creamos la ruta para exportar el df a un archivo parquet
ruta_parquet = 'sentiment.parquet'
```

```python
data_sentiment_grouped.to_parquet(ruta_parquet, engine='fastparquet')
```
